```python
for col in df.columns:
    print(df[col].nunique(), end = '\t:')
    print(col)
```

```python
df_pr = df[df['Country Name'] == 'Arab World']
```

```python
data = df_pr.values

for i in range(1, len(data)):
    print(data[i], data[i-1])
```

```python
data = df_pr.values

for i in range(1, len(data)):
    
    prev_year = data[i-1][3]
    curr_year = data[i][3]
#     print(data[i][3] - data[i-1][3])
    print(data[i][2], ':', round(((curr_year - prev_year)/prev_year)*100, 2), '%')
```

```python
df_pr = df[df['Country Name'] == 'India']

data = df_pr.values

gdp_change = [0]

for i in range(1, len(data)):
    gdp_change.append(round(((data[i][3] - data[i-1][3]) / data[i-1][3]) * 100, 2))
    
df_pr.assign(GDP = gdp_change)
```
